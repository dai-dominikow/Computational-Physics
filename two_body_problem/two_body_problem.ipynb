Conocida por ser 'la estrella mas brillante del cielo terrestre', es en realidad un sistema binario de dos estrellas:
- Sirius A con una masa $ m_a = 3.978x10^{30} kg $ y un radio $ r_a = 1.190x10^{6} km $<br>
- Sirius B con una masa $ m_b = 2.025x10^{30} kg $ y un radio $ r_b = 5843.9 km $ <br>

La distancia que separa estas estrellas es de $ r = 3x10ˆ{9}km $ <br>

<h2>Movimiento del sistema, demostración.</h2> 

Hay dos maneras de modelar el sistema: con el movimiento del sistema respecto al origen, y con el movimiento del sistema respecto al centro de masa.<br>
Para hacerlo, debemos reducir el problema a su equivalente del problema de un cuerpo, ya que son cuerpos regulates y trabajaremos con el centro de masa del sistema como referencia.<br>

En principio, tenemos 6 grados de libertad:<br>
\begin{equation}
\vec{r_1}=<x_1, y_1, z_1> 
\end{equation}

\begin{equation}
\vec{r_2}=<x_2, y_2, z_2>
\end{equation}

Lo primero que debemos obtener es el vector de la posición relativa $\vec{r}$<br>
\begin{equation}\vec{r} = \vec{r_2} - \vec{r_1}\end{equation}

Por otro lado, necesitamos también el vector $\vec{R}$ de la posición del centro de masa <br>
\begin{equation}
\vec{R} = \frac{m_1 \vec{r_1} + m_2 \vec{r_2}}{m_1 + m_2} = \frac{m_1 \vec{r_1} + m_2 \vec{r_2}}{M}
\end{equation}

Con esta información, reemplazamos en el Lagrangiano, que se define como la diferencia entre la energía cinética del sistema $T$  y su energía potencial $U$ <br>
$L = T - U$ y depende únicamente de $ r = | \vec{r_2} - \vec{r_1} |$
La energía cinética entonces <br>
\begin{equation}
T = \frac{1}{2}m_1(\dot{x_1^2}+\dot{y_1^2}+\dot{z_1^2}) + \frac{1}{2}m_2(\dot{x_2^2}+\dot{y_2^2}+\dot{z_2^2})
\end{equation}

\begin{equation}
T = \frac{1}{2}m_1|\dot{\vec{r}}^2_1| + \frac{1}{2}m_1|\dot{\vec{r}}^2_2|
\end{equation}

Voy a hacer un cambio de coordenadas y expresar $\vec{r_1}$ como $\vec{r_2} - \vec{r}$<br>

\begin{equation}
\vec{R} = \frac{m_1(\vec{r_2}-\vec{r}) + m_2 \vec{r_2}}{M} = \frac{(m_1 +m_2 )\vec{r_2} + m_1 \vec{r}}{M} = \vec{r_2} - \frac{m_1\vec{r}}{M}
\end{equation}
Por lo tanto
\begin{equation}
\vec{r_2} =  \vec{R} + \frac{m_1\vec{r}}{M}
\end{equation}
Y, de manera análoga
\begin{equation}
\vec{r_1} =  \vec{R} - \frac{m_2\vec{r}}{M}
\end{equation}

Siguiente paso, es derivar estos vectores<br>
\begin{equation}
\dot{\vec{r_2}} = \dot{\vec{R}}  +  \frac{m_1 \dot{\vec{r}}}{M}
\end{equation}

\begin{equation}
\dot{\vec{r_1}} = \dot{\vec{R}}  -  \frac{m_2 \dot{\vec{r}}}{M}
\end{equation}

y reemplazar todo en $T$<br>
\begin{equation}
T = \frac{1}{2}m_1(\dot{R} - m_2 \dot{r})^2 + \frac{1}{2}m_1(\dot{R} + m_1 \dot{r})^2|
\end{equation}
resueltos los cuadrados, y cancelados los terminos de primer orden obtenemos<br>
\begin{equation}
T = \frac{1}{2}M\dot{R}^2+ \frac{1}{2}\mu\dot{r}^2
\end{equation}

Observemos que el primer término es la energía cinética respecto al origen del sistema, y el segundo respecto al centro de masa.<br>
Volviendo al lagrangiano, y reemplazando por el resultado de la energía cinética, podemos obtener la siguiente expresión.
\begin{equation}
L = \frac{1}{2}M\dot{R}^2+ \frac{1}{2}\mu\dot{r}^2 - U(r)
\end{equation}
Donde todavía no definimos la energía potencial pero sabemos que es función de $\vec{r}$, a su vez, podemos determinar que $\dot{R}$ es una coordenada cíclica, es decir, que el lagrangiano no contiene a R y el centro de masa está fijo, o se mueve con una velocidad constante. Finalmente, podemos descibir nuestro sistema binario en coordenadas polares, sabiendo que el momento angular del mismo será constante.<br>

\begin{equation}
\vec{P}= M\dot{\vec{R}} = cte
\end{equation}

\begin{equation}
\vec{R} = \vec{R_0} + \frac{\vec{P}}{M}t
\end{equation}

De donde podemos obtener la velocidad del centro de masa.
\begin{equation}
\vec{R} = \vec{R_0} + \vec{V}t
\end{equation}

Pero, como vamos a analizar el sistema del centro de masas $\vec{P_0}=0$,$\vec{R}=0$<br>
Por lo tanto, el movimiento de las estrellas en torno al centro de masa es 
\begin{equation}
L = \frac{\mu}{2}\dot{\vec{r}}^2 - U(r)
\end{equation}

Y haciendo el cambio a coordenadas polares, nuestras derivadas  quedan:
\begin{equation}
\dot{\vec{r_x}} = \dot{r} \cos{\theta} - r \dot{\theta} \sin{\theta}
\end{equation}

\begin{equation}
\dot{\vec{r_y}} = \dot{r} \sin{\theta} + r \dot{\theta} \cos{\theta}
\end{equation}

\begin{equation}
\dot{\vec{r_x}}^2 = \dot{r}^2 \cos^{2}{\theta} - 2 r \dot{r}\dot{\theta}\sin{\theta}\cos{\theta} + r^2 \dot{\theta}^2 \sin^{2}{\theta}
\end{equation}

\begin{equation}
\dot{\vec{r_y}}^2 = \dot{r}^2 \sin^{2}{\theta} + 2 r \dot{r}\dot{\theta}\sin{\theta}\cos{\theta} - r^2 \dot{\theta}^2 \cos^{2}{\theta}
\end{equation}

Observemos que al sumar miembro a miembro estas derivadas para obtener $\dot{r}^2$ los términos no cuadráticos se anulan, y aplicando la identidad $\cos^{2}{\theta} + \sin^{2}{\theta} = 1$ finalmente podemos decir que 
\begin{equation}
(\dot{r_x}^2 + \dot{r_y}^2) =  \dot{r}^2 + r^{2}\dot{\theta}^2
\end{equation}

Y la energía cinética del lagrangiano queda expresada como
\begin{equation}
T  = \frac{1}{2}\mu(\dot{r}^2 + r^{2}\dot{\theta}^2)
\end{equation}

Veamos que nuevamente encontramos una coordenada cíclica $\dot{\theta}$, entonces, el momento del sistema se conserva.
\begin{equation}
P_{\theta} = \frac{\partial L}{\partial \dot{\theta}} = \mu r^2 \dot{\theta} = l
\end{equation}

El momento $P_{\theta}$ es una primer integral del movimiento y es igual a la magnitud del momento angular constante.<br>
La masa reducida se mueve un $d\theta$ en un tiempo $dt$, el área $d A$ recorrida por este radiovector $r$ está dada por
\begin{equation}
dA = \frac{1}{2} r^2  d\theta =  \frac{1}{2} r^2  \theta dt
\end{equation}

Nos queda entonces la 2da ley de Kepler
\begin{equation}
\frac{dA}{dt} = \frac{l}{2\mu} = cte
\end{equation}

Reemplazando T en el lagrangiano 
\begin{equation}
L  = \frac{1}{2}\mu(\dot{r}^2 + r^{2}\dot{\theta}^2) - U(r)
\end{equation}

\begin{equation}
\frac{\partial L}{\partial r}  = \mu r +\dot{\theta}^2 - \frac{\partial U}{\partial r}
\end{equation}

\begin{equation}
\frac{\partial L}{\partial \dot{r}}  = \mu\dot{r} 
\end{equation}

\begin{equation}
\frac{\partial }{\partial \dot{t}} \frac{\partial L}{\partial \dot{r}}  =\frac{\partial }{\partial \dot{t}} \mu\dot{r}  = \mu \ddot{r}
\end{equation}

Reemplazo en la primer derivada del lagrangiano
\begin{equation}
\mu \ddot{r} =\mu r \dot{\theta} - \frac{\partial U}{\partial r} 
\end{equation}

Con la expresión del momento angular y despejando $\dot{\theta}$ obtengo que
\begin{equation}
\mu \ddot{r} =\mu r (\frac{l}{\mu r^2})^2  - \frac{\partial U}{\partial r} 
\end{equation}

Llegando finalmente a un grado de libertad.
\begin{equation}
\mu \ddot{r} =\frac{l^2}{\mu r^3}  - \frac{\partial}{\partial r} U(r)
\end{equation}

Donde el primer término corresponde a la fuerza centrífuga $F_c$, y el segundo a la fuerza real
\begin{equation}
F = \frac{G m_1 m_2 \check{r}}{|r|^2}
\end{equation}

Finalmente, y para obtener la energía potencial efectiva del sistema
\begin{equation}
U_{ef} = U(r) + U_c = U(r) + \int - F_c dr = U(r) +  \frac{l^2}{2\mu r^2}
\end{equation}

$U(r)$ es el potencial gravitatorio.
\begin{equation}
U(r) = U_g = - \frac{G m_1 m_2}{|\vec{r}|}
\end{equation}

<h2> Ya con todo esto, podemos finalmente pasar a la simulación dinámica.</h2>
Para ello, necesito el momento inicial de uno de los dos cuerpos, ya que, por conservación del momento

\begin{equation}
P_{Sirius A} = - P_{Sirius B} 
\end{equation}

\begin{equation}
\vec{F_c} = m_1 \vec{a_x}
\end{equation}

\begin{equation}
|\vec{a}| = \frac{v^2}{r} = \omega^2 r
\end{equation}

\begin{equation}
m_1 \vec{a_x} = G\frac{m_1 m_2}{r^2}
\end{equation}

\begin{equation}
\vec{a_x} = G\frac{ m_2}{r^2}
\end{equation}

\begin{equation}
v_1 = \sqrt{\frac{ Gm_2}{r^2}r_1}
\end{equation}

Donde, si nos remitimos al principio de la demostración, depejamos que $r_1 = -\frac{m_2 \vec{r}}{M}$, por lo tanto:


\begin{equation}
v_1 = \sqrt{G\frac{ m_2}{r^2}-\frac{m_2 \vec{r}}{M}} = - \sqrt{G\frac{ m_2^2}{rM}}
\end{equation}





```python
import numpy as np
import matplotlib.pyplot as plt
import matplotlib.animation as anim
import pandas as pd
import math
```

```python
#Parámetros de mi sumulación, si quisiera simular otros dos cuerpos celestes, debo cambiarlos.
m1 = 3.978e30 #masa de SiriusA en kg
m2 = 2.025e30 #masa de SiriusB en kg
d = 3e9 #distancia entre las estrellas en m
tf = 24*3600*10  #tiempo de simulación en s
dt = 1000 #paso del tiempo en s
#Constantes, no van a cambiar durante la simulación
G = 6.6743e-11 #Nm^2/kg^2
```

```python
# debug 2, por ahora el mejor
def two_body_problem(m1, m2, d, tf, dt):
    #iterador
    i = 0        
    #tiempo simulación
    t = np.arange(0 , tf ,dt)
    #masa total
    M = m1+m2 
    #masa reducida
    mu = m1*m2/M  
    #posición centros de masa
    x1 = (m2/M)*d
    x2 = -(m1/M)*d

    #posicion inicial masas
    r1_i = np.array([d/2,0,0])
    r2_i = np.array([-d/2,0,0])
    #ahora, les tengo que dar un momento inicial a ambas estrellas
    #solo necesito la velocidad  de una de ellas
    v_1 =   np.sqrt((G*m2**2)/(d*M))
    #para una orbita circular
    v1_i = np.array([0,v_1,0])
    #por conservacion del momento
    p1_i= m1*v1_i
    p2_i=-p1_i
    #donde van a caer las iteraciones
    r1 = np.zeros((len(t),3))
    r1[i] = r1_i
    r2 = np.zeros((len(t),3))
    r2[i] = r2_i
    v1 = np.zeros((len(t),3))
    v1[i] = v1_i
    v2 = np.zeros((len(t),3))
    v2[i] = p2_i/m2
    #norma del vector r
    rnorm = np.zeros(len(t))
    rnorm[i] = np.linalg.norm(r2[i]-r1[i])
    #momento angular
    l = np.linalg.norm(np.cross(r1_i,p1_i) + np.cross(r2_i,p2_i)) 
    #energía
    Ug = np.zeros(len(t))
    Ug[i] = -G*m1*m2/d
    Uc = np.zeros(len(t))
    Uc[i] =  l**2/(2*mu*rnorm[i]**2)


    while i < len(t)-1:
        r_unitario = (r2[i]-r1[i])/rnorm[i]
        #obtengo la  magnitud de la fuerza
        F = G*m1*m2*r_unitario/rnorm[i]**2
        #velocidad y posición
        v1[i+1] =v1[i] + F/m1 * dt
        v2[i+1] =v2[i] - F/m2 * dt
        r1[i+1] =r1[i] + (v1[i+1]*dt)
        r2[i+1] =r2[i] + (v2[i+1]*dt)
        rnorm[i+1] =np.linalg.norm(r2[i+1]-r1[i+1])
        l = np.linalg.norm(np.cross(r1[i+1],v1[i+1]*m1) + np.cross(r2[i+1],v2[i+1]*m2))  
        Ug[i+1] =  - G*m1*m2/(rnorm[i+1])
        Uc[i+1] = l**2/(2*mu*rnorm[i+1]**2) 
        i+=1 
    return t, v1,v2,r1,r2,rnorm, Ug+Uc , Ug, Uc

t, v1,v2,r1,r2,rnorm,Uef, Ug, Uc = two_body_problem(m1, m2, d, tf, dt)  
```

```python
#### Análisis de resultados
ax = plt.figure(figsize=(6,6)).add_subplot( projection='3d',aspect = 'equal'  )
ax.plot(r1[:,0],r1[:,1],r1[:,2],color = 'lime', label = 'Sirius A')
ax.plot(r2[:,0],r2[:,1],r2[:,2],color = 'red', label = 'Sirius B')

ax.set_title('Posición respecto al centro de masa')
plt.show()

figure, axis = plt.subplots(2,2,figsize=(12,12))

axis[0][0].plot(np.linalg.norm(r1,axis = 1), np.linalg.norm(v1,axis = 1),  color = 'lime', label = 'Sirius A')
axis[0][0].plot(np.linalg.norm(r2,axis = 1), np.linalg.norm(v2,axis = 1),  color = 'tomato', label = 'Sirius B')
axis[0][0].set_xlabel("r")   
axis[0][0].set_ylabel("v")   
axis[0][0].set_title("Diagrama de fases")   
axis[0][0].legend(loc='upper left')


axis[0][1].plot( t,np.linalg.norm(v1,axis = 1),  color = 'lime', label = 'Sirius A')
axis[0][1].plot(t,np.linalg.norm(v2,axis = 1), color = 'red', label = 'Sirius B')
axis[0][1].set_xlabel("Tiempo")   
axis[0][1].set_ylabel("Velocidad")   
axis[0][1].set_title("Evolución velocidad de los cuerpos")   
axis[0][1].legend(loc='upper left')

axis[1][0].plot(rnorm, Uef,  color = 'pink', label = 'Uef')
axis[1][0].set_xlabel("Distancia entre los cuerpos (m)")   
axis[1][0].set_ylabel("Energía efectiva (J)")   
axis[1][0].set_title("Energía efectiva del sistema respecto a la distancia")   
axis[1][0].legend(loc='upper left')

axis[1][1].plot( t,rnorm,  color = 'teal', label = '|r|')
axis[1][1].set_xlabel("Tiempo")   
axis[1][1].set_ylabel("Distancia entre los cuerpos (m)")   
axis[1][1].set_title("Evolución distancia entre los cuerpos")   
axis[1][1].legend(loc='upper left')
plt.show()

```
